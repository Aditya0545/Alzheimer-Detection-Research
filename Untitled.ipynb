```python
RANDOM_STATE = 42
sns.set()
```

```python
df_model = df.drop(columns=['PatientID', 'DoctorInCharge'])
```

```python
df_model.tail()
```

```python
X = df_model.drop(columns=['Diagnosis'])
y = df_model['Diagnosis']
```

```python
non_numeric = X.select_dtypes(include=['object']).columns.tolist()
if len(non_numeric) > 0:
    print("Non-numeric columns found (will be dropped):", non_numeric)
    X = X.drop(columns=non_numeric)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
```

```python
# Scale numeric features (important for SVM and KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
joblib.dump(scaler, "scaler.joblib")
```

```python
models = {
    "SVM": SVC(kernel='linear', probability=True, class_weight='balanced', random_state=RANDOM_STATE),
    "RandomForest": RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=RANDOM_STATE),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "DecisionTree": DecisionTreeClassifier(class_weight='balanced', random_state=RANDOM_STATE)
}
```

```python
voting = VotingClassifier(
    estimators=[('svm', models['SVM']), ('rf', models['RandomForest']),
                ('knn', models['KNN']), ('dt', models['DecisionTree'])],
    voting='soft'  # use probabilities
)

models['VotingEnsemble'] = voting
```

```python
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    cm = confusion_matrix(y_test, y_pred)

    results[name] = {
        "Accuracy": acc,
        "Precision": prec,
        "F1 Score": f1,
        "Confusion Matrix": cm
    }

    print(f"\n{name} Results:")
    print("Accuracy:", acc)
    print("Precision:", prec)
    print("F1 Score:", f1)
    print("Confusion Matrix:\n", cm)
    print("Classification Report:\n", classification_report(y_test, y_pred, zero_division=0))
```

```python
# 5. Compare Results
# -------------------------------------------
results_df = pd.DataFrame(results).T[['Accuracy','Precision','F1 Score']]
print("\nModel Comparison:\n", results_df)

plt.figure(figsize=(10,6))
results_df[['Accuracy','Precision','F1 Score']].plot(kind='bar')
plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.show()
```

```python
best_model = models["Voting Ensemble"]
joblib.dump(best_model, "alzheimers_model.pkl")
print("\nâœ… Model saved as alzheimers_model.pkl")
```
